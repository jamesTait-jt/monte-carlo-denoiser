```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
itrs = 300
lr = 0.06

xs = [2] + [0] * (itrs - 1)
ys = [2] + [0] * (itrs - 1)
xys = [0] * itrs

for i in range(itrs - 1):
    xs[i + 1] = xs[i] - ys[i] * lr
    ys[i + 1] = ys[i] + xs[i] * lr
    xys[i] = xs[i] * ys[i]  
```

```python
#Plot the curve

fig = plt.figure()
plt.plot(xs, label="x")
plt.plot(ys, label="y", color="r")
plt.plot(xys, label="xy", linestyle="dotted", color="g")

plt.xlabel("Training iterations")
plt.legend()

plt.savefig("nash_failure.png")
plt.show()
```
